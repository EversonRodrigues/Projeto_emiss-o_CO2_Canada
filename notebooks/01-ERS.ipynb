```python
import re

def string_to_snake_case(string):
    string = re.sub("/", " ", string)                       # substitui / por espao
    string = re.sub(" +", " ", string)                      # substitui mltiplos espaos por um espao
    string = re.sub(" ", "_", string)                       # substitui espao por _
    string = re.sub("([a-z0-9])([A-Z])", r"\1_\2", string)  # adiciona _ entre letras minsculas e maisculas
    string = re.sub(r"\(", "", string)                      # elimina parnteses
    string = re.sub(r"\)", "", string)                      # elimina parnteses
    return string.lower()                                   # retorna a string em minsculo
```

```python
[string_to_snake_case(coluna) for coluna in df.columns]
```

```python
df.columns = [string_to_snake_case(coluna) for coluna in df.columns]

df.head()
```
